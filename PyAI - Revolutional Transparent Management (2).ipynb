# The Opportunity

Wealth managers are under pressure from two sides.
Clients want clear, story driven updates they can actually understand. Regulators, on the other hand, are demanding more detail, evidence, and transparency<sup>[1]</sup><sup>[2]</sup>.
In the UK, new Consumer Duty rules require firms to prove that their communications genuinely help customers make sense of their portfolios. When markets turned chaotic during the 2022 gilt crisis, it became obvious how much investors need clear explanations of what went wrong and why<sup>[3]</sup>.
Now, with global wealth bouncing back and digital programs restarting, firms are once again investing in technology to rebuild trust. The problem is that producing compliant, client-ready reports is still painfully manual<sup>[9]</sup>. Teams spend hours reconciling spreadsheets or validating data or also documenting audit trails. For smaller firms that effort eats directly into profit.
Moreover, the rapid adoption of AI and data-driven outcome in the financial sector is setting higher and new expectations among both clients and regulators. In the past it was once considered as in "innovation", but now is turning into a minimum standard to remain competitive. Wealth managers who are unable to communicate clearly risk are left behind by peers who can provide both clarity and accuracy. 
The result is a clear need: faster and more transparent reporting that turns data into human readable insight, without adding to the workload. 




# Our Solution - The PyAI Platform for Hedge Funds*

For hedge funds and their end investors (direct clients or via mandates), PyAI ingests positions, transactions and also NAV/admin files, then decomposes performance and surfaces investment movements in a client dashboard that is <span style="color:#3fa9f5;font-weight:600;">clear</span> and <span style="color:#3fa9f5;font-weight:600;">fully</span> traceable. The goal is to give investors the near real‑time visibility they expect while automating report production so teams no longer edit letters or packs by hand every month.
This approach connect the traditional gap between data analytics and investor communication, making sure that performance metrics are not only proper but also meaningful. PyAI assists as both a governance layer and a storytellin tool,translating dense data outcomes into intuitive insights that fortify transparency and confidence mainly for the clients.  

## The movements dashboard (investor‑facing)
- **<span style="color:#3fa9f5;font-weight:600;">Flows & capital</span>:** subscriptions/redemptions by period, net cash flows, dealing dates, redemption frequency, lock‑up/notice periods; gate alerts where applicable.  
- **<span style="color:#3fa9f5;font-weight:600;">Portfolio movements</span>:** trade/rotation timeline (opens/closes, add‑ons) with thesis tags (e.g., “value”, “merger arb”, “macro rates”).  
- **<span style="color:#3fa9f5;font-weight:600;">Exposures & risks</span>:** gross/net exposure by strategy (L/S Equity, Credit, Macro, Commodity...), sensitivities (equity beta, duration, FX), VaR/drawdown, and scenario stress tests.

## How it works (operating flow)
Our Python program <span style="color:#3fa9f5;font-weight:600;">automatically collects</span> the data the hedge funds already produce (prime‑broker statements, fund‑administrator files (NAV/share classes), trade blotters and PMS/OMS exports) via CSV/SFTP/API and keeps dated copies so every figure is traceable. With this structure, PyAI guarantees that every reported figure can be recorded back to its origin. This way of making the audit ready and transparent not only satisfies compliance but also helps to biuld a long-term credit with institutional investors who more and more ask for verifiable reporting pipelines.

# Challenges & Mitigation

## Model reliability & governance
**The challenge** When AI helps write or analyze client materials, regulators expect tight control: clear ownership, testing, monitoring and sign-off. The EU AI Act will add more transparency rules <sup class="ref">[4]</sup><sup class="ref">[5]</sup>.
**What we do** We test models against clear quality metrics, keep a human reviewer in the loop for all client wording, and track every change. Every line of text links back to the exact numbers and charts, so the audit trail is solid. We also keep an ongoing validation scheme to keep track of drift in model performance to ensure that updates do not bring inconsistencies or bias in client-view outputs. 

## Outsourcing / third-party risk
**The challenge** If any service is external, supervisors want contracts, resilience plans, exit routes, incident notices, and clarity on where data lives (PRA SS2/21)<sup class="ref">[10]</sup>.
**What we do** We use standard data contracts and SLAs, map dependencies, prepare exit/transition playbooks, keep incident runbooks with contact windows, and let clients choose data residency. These safeguards create an additional stage of operational strength and provide clients safety that PyAI can operate securely within the strictest institutional frameworks. 

## Communication outcomes
**The challenge** Under the UK Consumer Duty, firms must prove that clients really understand the message—not just that they received a disclosure <sup class="ref">[1]</sup>.
**What we do** Our reports follow a simple flow: *what happened* and *why it happened* and what is the consequence for the client with for example A/B tests and readability checks. All claims point back to the data, so understanding can be evidenced. This way of work that is base on feedbacks ensures that communication becomes measurable; in this way, firms can over time continue to imrpove their reports based on real comprehension figures rather than expectations.

```python
# =================== PyAI Monthly Report (Donut + Drift/Trade, client-friendly) ===================
# One cell: prompts → Yahoo fetch → M-1 vs M math → visuals + plain-English explanations

import sys, warnings
from datetime import date, timedelta

def _pip_install(p):
    import subprocess, sys
    try:
        subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", p])
    except Exception as e:
        print(f"!! pip install {p} failed: {e}")

# Deps
try:    import pandas as pd
except: _pip_install("pandas"); import pandas as pd
try:    import numpy as np
except: _pip_install("numpy");  import numpy as np
try:    import yfinance as yf
except: _pip_install("yfinance"); import yfinance as yf

import matplotlib.pyplot as plt
from matplotlib import patheffects as pe
from IPython.display import display, HTML

# ---- Brand look (PyAI)
NAVY  = "#0b2545"
LIGHT = "#3fa9f5"
GRAY  = "#475569"

# ---------------- Month helpers ----------------
def last_full_month_label():
    today = date.today().replace(day=1)
    end_last = today - timedelta(days=1)
    return f"{end_last.year:04d}-{end_last.month:02d}"

def parse_month_label(mstr):
    y, m = mstr.split("-"); y, m = int(y), int(m)
    assert 1 <= m <= 12
    return y, m

def month_bounds(y, m):
    start = date(y, m, 1)
    end = (date(y+1,1,1)-timedelta(days=1)) if m==12 else (date(y,m+1,1)-timedelta(days=1))
    return start, end

def prev_month(y, m):
    return (y-1, 12) if m==1 else (y, m-1)

# ---------------- Inputs ----------------
def parse_weights(s):
    """Accepts: 'AAPL:20, MSFT 30, RACE 50%'. Returns weights normalized to 1.0."""
    s = (s or "").replace(";", ",")
    items = [p.strip() for p in s.split(",") if p.strip()]
    out = {}
    for p in items:
        if ":" in p:
            t, w = p.split(":", 1)
        else:
            toks = p.split()
            if len(toks) < 2: continue
            t, w = " ".join(toks[:-1]), toks[-1]
        t = t.strip().upper()
        w = w.strip().replace("%","")
        out[t] = float(w)
    if not out: raise ValueError("No valid ticker:weight pairs.")
    total = sum(out.values())
    scale = 1.0 if total <= 1.5 else 100.0
    out = {k: v/scale for k,v in out.items()}
    tot = sum(out.values())
    out = {k: v/tot for k,v in out.items()}
    return out

# ---------------- Yahoo helpers ----------------
def fetch_close(tickers, start, end):
    """Auto-adjusted Close with robust column handling."""
    end_inc = (end + timedelta(days=2)).isoformat()
    with warnings.catch_warnings():
        warnings.simplefilter("ignore")
        df = yf.download(tickers, start=start.isoformat(), end=end_inc,
                         auto_adjust=True, progress=False, threads=True)
    if df is None or len(df)==0:
        raise RuntimeError("Yahoo returned no data.")
    if isinstance(df.columns, pd.MultiIndex):
        if "Close" in df.columns.get_level_values(0): close = df["Close"]
        elif "Adj Close" in df.columns.get_level_values(0): close = df["Adj Close"]
        else: raise RuntimeError("Neither Close nor Adj Close present.")
    else:
        close = df.copy()
    # map dash/dot alternative symbols if needed
    for t in tickers:
        if t not in close.columns:
            alt = t.replace("-", ".") if "-" in t else t.replace(".", "-")
            if alt in close.columns: close = close.rename(columns={alt: t})
    return close

def last_trading_close_by_month(close_df, y, m):
    """Return the last trading day's close for month (y,m)."""
    mask = (close_df.index.year==y) & (close_df.index.month==m)
    sub = close_df.loc[mask].dropna(how="all")
    if sub.empty:
        before = close_df.loc[:date(y, m, 1)]
        if before.empty: raise RuntimeError(f"No prices for {y}-{m:02d}.")
        return before.iloc[-1]
    return sub.ffill().iloc[-1]

# ---------------- Company metadata ----------------
def safe_info(t):
    try: info = yf.Ticker(t).get_info() or {}
    except Exception: info = {}
    return {
        "sector":  info.get("sector")  or "Unknown",
        "industry":info.get("industry")or "Unknown",
        "country": info.get("country") or "Unknown",
        "marketCap": info.get("marketCap")
    }

def cap_bucket(cap):
    if cap is None or (isinstance(cap,float) and np.isnan(cap)): return "Unknown"
    if cap >= 200e9: return "Mega"
    if cap >= 10e9:  return "Large"
    if cap >= 2e9:   return "Mid"
    if cap >= 0.3e9: return "Small"
    return "Micro"

# ---------------- Donut utilities ----------------
def topN_with_other(weights_pct, N=8, label_other="Other"):
    s = weights_pct.dropna().sort_values(ascending=False)
    if len(s) <= N: return s
    top = s.iloc[:N].copy()
    other = s.iloc[N:].sum()
    if other > 0: top.loc[label_other] = other
    return top

def donut_palette(n):
    from matplotlib.colors import to_rgb
    a = np.array(to_rgb(LIGHT)); b = np.array(to_rgb(NAVY))
    t = np.linspace(0, 1, n)
    return [tuple(a*(1-x) + b*x) for x in t]

def plot_donut(series, title):
    """Doughnut chart with white percentages inside the ring."""
    s = topN_with_other(series, N=8)
    colors = donut_palette(len(s))
    fig, ax = plt.subplots(figsize=(6.2, 4.8))
    wedges, texts, autotexts = ax.pie(
        s.values,
        labels=s.index,
        labeldistance=1.05,       # labels around
        autopct="%.1f%%",         # percentages
        pctdistance=0.78,         # % inside the ring
        startangle=90,
        colors=colors,
        textprops={"color": GRAY, "fontsize": 9},
        wedgeprops={"width": 0.45}
    )
    for t in autotexts:
        t.set_color("white")
        t.set_fontsize(9)
        t.set_weight("bold")
        t.set_path_effects([pe.withStroke(linewidth=1.0, foreground="black", alpha=0.15)])
    ax.set_aspect("equal")
    ax.set_title(title, color=NAVY, fontsize=13)
    plt.tight_layout(); plt.show(); plt.close(fig)

# ---------------- Explanations ----------------
def explain_concentration(series, dim):
    s = series.sort_values(ascending=False)
    top = list(s.head(3).items())
    top_txt = "; ".join([f"{k}: {v:.1f}%" for k,v in top])
    msg = (f"The portfolio is currently tilted by <b>{dim.lower()}</b> toward <b>{top[0][0]}</b> "
           f"({top[0][1]:.1f}%), with additional exposure in {', '.join([k for k,_ in top[1:]])}. ")
    if "Other" in s.index:
        small = series.drop("Other")
        tail_count = (small < 3).sum()
        if tail_count >= 3:
            msg += f"A long tail of {tail_count} smaller buckets (<3% each) adds diversification. "
    msg += ("These percentages show the share of the total portfolio invested in each category, "
            "so a non-specialist can see the high-level stance at a glance.")
    return msg

def explain_change(delta_pp, dim):
    inc = delta_pp.sort_values(ascending=False)
    dec = delta_pp.sort_values(ascending=True)
    up  = f"<b>{inc.index[0]}</b> (+{inc.iloc[0]:.2f}pp)"
    dn  = f"<b>{dec.index[0]}</b> ({dec.iloc[0]:.2f}pp)"
    return (f"By <b>{dim.lower()}</b>, the largest increase was {up}; the largest reduction was {dn}. "
            "Positive numbers mean a bigger slice at month-end, either because prices rose (<i>drift</i>) "
            "or because we added positions (<i>trade</i>).")

def explain_contrib(contrib_pct, dim):
    s = contrib_pct.sort_values(ascending=False)
    pos = f"<b>{s.index[0]}</b> (+{s.iloc[0]:.2f}%)"
    neg = f"<b>{s.index[-1]}</b> ({s.iloc[-1]:.2f}%)"
    return (f"By <b>{dim.lower()}</b>, performance was led by {pos}, while the main drag came from {neg}. "
            "Contribution = starting weight × return over the month.")

# =========================== PROMPTS ===========================
print("Enter tickers & weights at M-1 (e.g., AAPL:20, MSFT 30, RACE 50)")
w0_txt = input("M-1 weights: ").strip()
print("\nEnter tickers & weights at M   (same format; omitted tickers => 0%)")
w1_txt = input("M   weights: ").strip()
m_label = input("\nEnter M as YYYY-MM (blank = last full month): ").strip()
if not m_label: m_label = last_full_month_label()
yM, mM = parse_month_label(m_label); yM1, mM1 = prev_month(yM, mM)
print(f"→ Analysing M-1={yM1}-{mM1:02d} vs M={yM}-{mM:02d}")

# =========================== DATA & MATH ===========================
start_fetch = month_bounds(yM1, mM1)[0]
nextM_y, nextM_m = (yM+1,1) if mM==12 else (yM,mM+1)
end_fetch   = month_bounds(nextM_y, nextM_m)[0]

w0 = parse_weights(w0_txt); w1 = parse_weights(w1_txt)
tickers = sorted(set(w0)|set(w1))
close = fetch_close(tickers, start_fetch, end_fetch)
P_M1 = last_trading_close_by_month(close, yM1, mM1).reindex(tickers)
P_M  = last_trading_close_by_month(close, yM,  mM).reindex(tickers)

ret = (P_M/P_M1 - 1.0).fillna(0.0)
w0s = pd.Series({t:w0.get(t,0.0) for t in tickers})
w1s = pd.Series({t:w1.get(t,0.0) for t in tickers})

Rp = float((w0s*ret).sum())
passive_w1 = (w0s*(1+ret))/(1+Rp)
trade_w = (w1s - passive_w1).fillna(0.0)  # active buy/sell
drift_w = (passive_w1 - w0s).fillna(0.0)  # performance effect

# Metadata & groupings
meta = {t: safe_info(t) for t in tickers}
sector   = pd.Series({t:(meta[t]["sector"] or "Unknown")   for t in tickers})
industry = pd.Series({t:(meta[t]["industry"] or "Unknown") for t in tickers})
country  = pd.Series({t:(meta[t]["country"] or "Unknown")  for t in tickers})
capcat   = pd.Series({t:cap_bucket(meta[t]["marketCap"])   for t in tickers})

def agg_exposure(keys, w0s, w1s, contrib):
    g = pd.DataFrame({
        "W(M-1)%": w0s.groupby(keys).sum()*100,
        "W(M)%":   w1s.groupby(keys).sum()*100
    })
    g["ΔWeight_pp"] = g["W(M)%"] - g["W(M-1)%"]
    g["Contribution_%"] = contrib.groupby(keys).sum()
    return g.sort_values("ΔWeight_pp", ascending=False)

ex_sector   = agg_exposure(sector,   w0s, w1s, w0s*ret*100)
ex_country  = agg_exposure(country,  w0s, w1s, w0s*ret*100)
ex_industry = agg_exposure(industry, w0s, w1s, w0s*ret*100)
ex_cap      = agg_exposure(capcat,   w0s, w1s, w0s*ret*100)

# =========================== REPORT ===========================
month_name = f"{yM}-{mM:02d}"; prev_name = f"{yM1}-{mM1:02d}"

display(HTML(f"<h3 style='color:{NAVY};margin-top:8px;'>I. Portfolio Position — current month (M)</h3>"))

# Current weights (M)
Wm = w1s*100.0
by_country  = Wm.groupby(country).sum().sort_values(ascending=False)
by_sector   = Wm.groupby(sector).sum().sort_values(ascending=False)
by_industry = Wm.groupby(industry).sum().sort_values(ascending=False)
by_cap      = Wm.groupby(capcat).sum().sort_values(ascending=False)

plot_donut(by_country,  "Weights by Country (M)")
display(HTML(f"<p style='color:{GRAY};font-size:13px;'>{explain_concentration(by_country, 'Country')}</p>"))

plot_donut(by_sector,   "Weights by Sector (M)")
display(HTML(f"<p style='color:{GRAY};font-size:13px;'>{explain_concentration(by_sector, 'Sector')}</p>"))

plot_donut(by_industry, "Weights by Industry (M)")
display(HTML(f"<p style='color:{GRAY};font-size:13px;'>{explain_concentration(by_industry, 'Industry')}</p>"))

plot_donut(by_cap,      "Weights by Market-Cap (M)")
display(HTML(f"<p style='color:{GRAY};font-size:13px;'>{explain_concentration(by_cap, 'Market-Cap')}</p>"))

# ---- II. Change in Portfolio ----
display(HTML(f"<h3 style='color:{NAVY};margin-top:24px;'>II. Change in Portfolio — M-1 → M</h3>"))

def bar_change(s, title):
    s = s.sort_values(ascending=False)
    colors = [LIGHT if v>=0 else NAVY for v in s.values]
    fig, ax = plt.subplots(figsize=(9.2, 4.2))
    ax.bar(s.index, s.values, color=colors)
    ax.set_title(title, color=NAVY, fontsize=13)
    ax.set_ylabel("ΔWeight (pp)", color=GRAY, fontsize=10)
    ax.grid(True, axis="y", alpha=0.3)
    ax.tick_params(axis='x', labelsize=9); ax.tick_params(axis='y', labelsize=9)
    plt.xticks(rotation=45, ha="right"); plt.tight_layout(); plt.show(); plt.close(fig)

# ΔWeights by Sector & Country
bar_change((ex_sector["W(M)%"]-ex_sector["W(M-1)%"]),  "ΔWeight by Sector (M-1 → M)")
display(HTML(f"<p style='color:{GRAY};font-size:13px;'>{explain_change((ex_sector['W(M)%']-ex_sector['W(M-1)%']), 'Sector')}</p>"))

bar_change((ex_country["W(M)%"]-ex_country["W(M-1)%"]), "ΔWeight by Country (M-1 → M)")
display(HTML(f"<p style='color:{GRAY};font-size:13px;'>{explain_change((ex_country['W(M)%']-ex_country['W(M-1)%']), 'Country')}</p>"))

# Drift vs Trade by Sector (Top 10 |ΔW|)
drift_sec = pd.Series(drift_w, index=tickers).groupby(sector).sum()*100.0
trade_sec = pd.Series(trade_w, index=tickers).groupby(sector).sum()*100.0
order = (drift_sec.abs() + trade_sec.abs()).sort_values(ascending=False).index[:10]

fig, ax = plt.subplots(figsize=(9.6, 4.6))
ax.bar(order, drift_sec.reindex(order), color=LIGHT, label="Performance drift")
ax.bar(order, trade_sec.reindex(order), bottom=drift_sec.reindex(order), color=NAVY, label="Active trade")
ax.set_title("ΔWeight decomposition by Sector (Top 10 |ΔW|)", color=NAVY, fontsize=13)
ax.set_ylabel("pp (percentage points)", color=GRAY, fontsize=10)
ax.legend(fontsize=9)
ax.grid(True, axis="y", alpha=0.3)
ax.tick_params(axis='x', labelsize=9); ax.tick_params(axis='y', labelsize=9)
plt.xticks(rotation=45, ha="right"); plt.tight_layout(); plt.show(); plt.close(fig)

display(HTML(f"<p style='color:{GRAY};font-size:13px;'>"
             "Reading guide: the light-blue bar shows weight change driven by performance (<i>drift</i>), "
             "and the navy segment shows extra change due to net buying/selling (<i>trade</i>). "
             "A large positive navy segment means exposure was actively increased; a large negative one means it was reduced.</p>"))

# Sector contribution
sec_contrib = ex_sector["Contribution_%"].sort_values(ascending=False)
fig, ax = plt.subplots(figsize=(9.0, 4.0))
ax.bar(sec_contrib.index, sec_contrib.values, color=[LIGHT if v>=0 else NAVY for v in sec_contrib.values])
ax.set_title("Return contribution by Sector (W(M-1) × r)", color=NAVY, fontsize=13)
ax.set_ylabel("Contribution (%)", color=GRAY, fontsize=10)
ax.grid(True, axis="y", alpha=0.3)
ax.tick_params(axis='x', labelsize=9); ax.tick_params(axis='y', labelsize=9)
plt.xticks(rotation=45, ha="right"); plt.tight_layout(); plt.show(); plt.close(fig)

display(HTML(f"<p style='color:{GRAY};font-size:13px;'>{explain_contrib(ex_sector['Contribution_%'], 'Sector')} "
             "For a quick read: positive bars boosted the overall portfolio return; negative bars reduced it.</p>"))

# Final wrap-up (expanded)
largest_up_sec = (ex_sector["W(M)%"]-ex_sector["W(M-1)%"]).idxmax()
largest_up_pp  = (ex_sector["W(M)%"]-ex_sector["W(M-1)%"]).max()
largest_dn_sec = (ex_sector["W(M)%"]-ex_sector["W(M-1)%"]).idxmin()
largest_dn_pp  = (ex_sector["W(M)%"]-ex_sector["W(M-1)%"]).min()

wrap = (f"<p style='color:{GRAY};font-size:13px;'>"
        f"Over the month, the portfolio returned <b>{Rp*100:.2f}%</b>. "
        f"The largest allocation increase was in <b>{largest_up_sec}</b> (+{largest_up_pp:.2f}pp) and the largest decrease in "
        f"<b>{largest_dn_sec}</b> ({largest_dn_pp:.2f}pp). "
        "Weights change for two reasons: (1) markets move, which passively shifts weights even without trading — "
        "that is <i>drift</i>; and (2) active decisions to buy or sell — that is <i>trade</i>. "
        "The doughnuts show today’s stance, the ΔWeight bars show how we got here, and the drift/trade view "
        "separates market effect from manager action so any reader can see what truly drove the change.</p>")
display(HTML(wrap))
```
